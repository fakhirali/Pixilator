```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
orig_image = cv2.imread("latest.png")
```

```python
plt.imshow(orig_image)
```

```python
def pixel(image, size):
    image = image.copy()
    width = (image.shape[1] // size ) + 1
    height = (image.shape[0] // size ) + 1 
    #print(height , width)
    for i in range(height):
        for j in range(width):
            r = np.average(image[i*size:(i*size)+size , j*size:(j*size)+size , 0])
            g = np.average(image[i*size:(i*size)+size , j*size:(j*size)+size , 1])
            b = np.average(image[i*size:(i*size)+size , j*size:(j*size)+size , 2])
            image[i*size:(i*size)+size , j*size:(j*size)+size] = [r,g,b]
    return image
```

```python
plt.imshow(pixel(orig_image,10))
```

```python
ten_by_ten = pixel(orig_image,10)
```

```python
cv2.imwrite('new.png',ten_by_ten)
```
